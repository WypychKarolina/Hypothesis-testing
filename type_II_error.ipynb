```python
import numpy as np
from scipy.stats import norm , chi2
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def second_mistake_zad1(mu,alternative = 'lt',test_mu = 1.5,alpha = 0.05, n=1000):
    sigma = 0.2
    sum = 0
    if alternative == 'lt':
        #H1: mu < 1.5 
        for _ in range(1000):
            X = norm.rvs(mu,sigma,n)
            Z = (np.mean(X)  - test_mu)/ sigma * np.sqrt(n)
            if Z >= norm.ppf(alpha):
                sum += 1
    elif alternative == 'gt':
        #H1: mu > 1.5 
        for _ in range(1000):
            X = norm.rvs(mu,sigma,n)
            Z = (np.mean(X)  - test_mu)/ sigma * np.sqrt(n)
            if Z <= norm.ppf(1-alpha):
                sum += 1
    elif alternative == 'ne':
        #H1: mu != 1.5
        for _ in range(1000):
            X = norm.rvs(mu,sigma,n)
            Z = (np.mean(X)  - test_mu)/ sigma * np.sqrt(n)
            if Z <= norm.ppf(1-alpha/2) and Z >= norm.ppf(alpha/2):
                sum += 1
    return sum/1000

```

```python
mu = np.linspace(1.47,1.53,20)
mistakes = []
for m in mu:
    sum = 0
    for _ in range(100):
        sum += second_mistake_zad1(m,'ne',n=1000)
    mistakes.append(sum/100)

plt.figure(figsize=(8, 5))
sns.lineplot(x=mu,y=mistakes,markers=True,label='Błąd  II rodzaju')
sns.lineplot(x=mu,y=1-np.array(mistakes),markers=True,label='Moc testu')
plt.legend()
plt.xlabel('$\mu$')
plt.grid()
plt.savefig('images/zad3_1a_II.pdf')
    
```

```python
def second_mistake_zad2(test_sigma2,alternative = 'lt',alpha = 0.05, n=1000):
    mu = 0.2
    sigma2 = 1.5
    sum = 0
    if alternative == 'lt':
        #H1: sigma ^2 < 1.5 
        for _ in range(1000):
            X = norm.rvs(mu,np.sqrt(test_sigma2),n)
            X2 = (n-1)*np.var(X)/sigma2
            if X2 >= chi2.ppf(alpha,n-1):
                sum += 1
    elif alternative == 'gt':
        #H1: sigma ^2 > 1.5 
        for _ in range(1000):
            X = norm.rvs(mu,np.sqrt(test_sigma2),n)
            X2 = (n-1)*np.var(X)/sigma2
            if X2 <= chi2.ppf(1-alpha,n-1):
                sum += 1
    elif alternative == 'ne':
        #H1: sigma ^2 != 1.5
        for _ in range(1000):
            X = norm.rvs(mu,np.sqrt(test_sigma2),n)
            X2 = (n-1)*np.var(X)/sigma2
            if X2 >= chi2.ppf(alpha/2,n-1) and X2 <= chi2.ppf(1-alpha/2,n-1):
                sum += 1
    return sum/1000
```

```python
sigmas2 = np.linspace(1.2,1.9,30)
mistakes = []
for sigma2 in sigmas2:
    sum = 0
    for _ in range(100):
        sum += second_mistake_zad2(sigma2,'ne',n=1000)
    mistakes.append(sum/100)

plt.figure(figsize=(8, 5))
sns.lineplot(x=sigmas2,y=mistakes,markers=True,label='Błąd  II rodzaju')
sns.lineplot(x=sigmas2,y=1-np.array(mistakes),markers=True,label='Moc testu')
plt.legend()
plt.xlabel('$\sigma^2$')
plt.grid()
plt.savefig('images/zad3_2a_II.pdf')
```
