```python
def test_1(alpha,hipo):
    k = 1000
    sigma=0.2
    mu = 1.5
    N=1000
    if hipo == "ne":
        row = 1-alpha/2
        quantile = scipy.stats.norm.ppf(row,0,1)
        X_vector = scipy.stats.norm.rvs(mu,sigma,(N,k))
        Z_test = (np.mean(X_vector,axis=1)-mu)/sigma*np.sqrt(k)
        answer=  (Z_test <= -quantile) + (Z_test >= quantile)
        return np.mean(answer)
    if hipo == "gt":
        row = 1-alpha
        quantile = scipy.stats.norm.ppf(row,0,1)
        X_vector = scipy.stats.norm.rvs(mu,sigma,(N,k))
        Z_test = (np.mean(X_vector,axis=1)-mu)/sigma*np.sqrt(k)
        answer= (Z_test >= quantile)
        return np.mean(answer)
    if hipo == "lt":
        row = 1-alpha
        quantile = scipy.stats.norm.ppf(row,0,1)
        X_vector = scipy.stats.norm.rvs(mu,sigma,(N,k))
        Z_test = (np.mean(X_vector,axis=1)-mu)/sigma*np.sqrt(k)
        answer= (Z_test <= -quantile)
        return np.mean(answer)
```

```python
M=100
repeat_mu_ne = [test_1(0.05,"ne") for i in range(M)]
```

```python
repeat_mu_gt = [test_1(0.05,"gt")for i in range(M)]
```

```python
repeat_mu_lt = [test_1(0.05,"lt")for i in range(M)]
```

```python
sns.boxplot(repeat_mu_ne)
plt.xlabel("Błąd I rodzaju")
plt.savefig("box1.pdf")
```

```python
def test_2(alfa,hipo):
    k = 1000
    mu = 1.5
    sigma=0.2
    zero_hipo = 1.5
    N=1000
    freedom = k-1
    if hipo == "ne":
        row_down = alpha/2
        row_up = 1-alpha/2
        quantile_down = scipy.stats.chi2.ppf(row_down,freedom)
        quantile_up = scipy.stats.chi2.ppf(row_up,freedom)
        X_vector = scipy.stats.norm.rvs(mu,sigma,(N,k))
        Chi_test = np.var(X_vector,axis=1)*freedom/sigma**2
        answer =  (Chi_test <= quantile_down) + (Chi_test >= quantile_up)
        return np.mean(answer)
    if hipo == "gt":
        row = 1-alpha
        quantile = scipy.stats.chi2.ppf(row,freedom)
        X_vector = scipy.stats.norm.rvs(mu,sigma,(N,k))
        Chi_test = np.var(X_vector,axis=1)*freedom/sigma**2
        answer = (Chi_test >= quantile)
        return np.mean(answer)
    if hipo == "lt":
        row = alpha
        quantile = scipy.stats.chi2.ppf(row,freedom)
        X_vector = scipy.stats.norm.rvs(mu,sigma,(N,k))
        Chi_test = np.var(X_vector,axis=1)*freedom/sigma**2
        answer = (Chi_test <= quantile)
    return np.mean(answer)
```
